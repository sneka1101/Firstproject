multiple subplot

```python
plt.subplot(1,2,1)  # (row,colum,plot)
plt.plot(x,y,'b--')
plt.subplot(1,2,2)
plt.plot(x,y,'g')
```

oops in matplotlib

```python

fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.5,0.5]) ##between 0 to 1 value

axes.plot(x,y,'g')
axes.set_xlabel('xaxis')
axes.set_label('yaxis')
axes.set_title('title')
```

create multiple plot using matplotlib

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,.8,.8])
axes2=fig.add_axes([.2,.5,.4,.3])

axes1.plot(x,y,'g')
axes1.set_xlabel('x-axis')
axes1.set_ylabel('y-axis')
axes1.set_title('title')


axes2.plot(x,y,'b--')
axes2.set_xlabel('xais')
axes2.set_ylabel('ylabel')
axes2.set_title('two plot')

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data with matching lengths
x = np.random.randint(0, 4, 7)  # 7 random integers between 0 and 4
y = np.random.randint(2, 6, 7)  # 7 random integers between 2 and 6

# Create scatter plot
plt.scatter(x, y, color='red')

# Adding title and labels
plt.title('Regenerated Random Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.show()
```

```python
x1=[2,4,5,7,8,90] #multiple plots
y1=[3,5,7,8,60,100]

x2=[10,20,30,40,90]
y2=[60,70,80,90,90]

plt.scatter(x1,y1,color='red')
plt.scatter(x2,y2,color='green')

plt.title('scatterplot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

Bar plot:

```python

sub=['communication','maths','social','dsp','science']
stud=[90,60,50,90,100]

plt.bar(sub,stud,color='green')

plt.title(' subject marks')
plt.xlabel('subject')
plt.ylabel('marks')
plt.show()
```

pie chart

```python
sub=['communication','maths','social','dsp','science']
stud=[90,60,50,90,100]

plt.pie(stud,labels=sub)
plt.show()
```

```python
plt.pie(stud,labels=sub, autopct= '%0.1f%%')#autopct is to define the percentage .1 it denotes after point one number
plt.show()
```

```python
plt.pie(stud,labels=sub , shadow=True)
plt.show()
```

```python
plt.pie(stud,labels=sub , shadow=True ,autopct= '%0.1f%%', radius = 1.6)#bigger size
plt.show()
```

```python
textprops={'fontsize':14,'color':'k'}

plt.pie(stud,labels=sub ,autopct= '%0.1f%%', textprops=textprops)
plt.show()
```

```python



plt.pie(stud,labels=sub ,autopct= '%0.1f%%')
plt.legend()
plt.show()
```

```python
colors={'red','green','blue','pink','grey'}
plt.pie(stud,labels=sub , shadow=True,colors=colors)
plt.show()
```
